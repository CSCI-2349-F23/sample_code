### Below I provide all the code from Class8.2 in one cell for Great Expectations.

```python
# all code in one place for Great Expectations

# open the file, read it in, replace newlines with space
f = open("greatexpectations.txt", encoding="utf=8")
fulltext = f.read()
alltext = re.sub("\n", " ", fulltext)

# remove the boilerplate
alltext = re.sub("^.*?Chapter I\.\s\s\s", "", alltext)
alltext = re.sub("\s+\*\*\* END OF THE.*?$", "", alltext)


# call nltk.word_tokenize
alltokens = nltk.word_tokenize(alltext)

# get stop word list
from nltk.corpus import stopwords
stoplist = stopwords.words('english')

# add important missing stuff to stop word list
stoplist.extend([".", ",", "?", "could", "would", "“", "”", "’", ";", "!","much", "like", "one", "many", "though", "without", "upon"])

# make a new version of alltokens called allcontenttokens that doesn't contain items from the stop list
allcontenttokens = [w for w in alltokens if w.lower() not in stoplist]

# lemmatize
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# create a new list of tokens, alllemmas by lemmatizing allcontenttokens
all_lemmas = [lemmatizer.lemmatize(w) for w in allcontenttokens]
```

### Here's the new code below for creating word clouds!

```python
# The wordcloud library expects a string not a list, 
# so join the list back together with spaces
text = " ".join(all_lemmas)

# Generate a word cloud image
wordcloud = WordCloud().generate(text)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

# lower max_font_size
wordcloud = WordCloud(max_font_size=40).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

### And here's the same code for Walden.

```python
# all code in one place for Walden

# read in file, replace "\n" with space
# and remove boilerplate

f = open("walden.txt", encoding="utf=8")
fulltext = f.read()
alltext = re.sub("\n", " ", fulltext)

alltext = re.sub("^.*?DISOBEDIENCE \*\*\*", "", alltext)
alltext = re.sub("\s+\*\*\* END OF THE.*?$", "", alltext)

# (note: this ^^ is different from Great Expectations because
# the beginning of the book is not Chapter I)

# call nltk.word_tokenize here 
alltokens = nltk.word_tokenize(alltext)

# make a new version of alltokens called allcontenttokens that doesn't contain items from the stop list
allcontenttokens = [w for w in alltokens if w.lower() not in stoplist]

# lemmatize
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# create a new list of tokens, alllemmas by lemmatizing allcontenttokens
all_lemmas = [lemmatizer.lemmatize(w) for w in allcontenttokens]

# The wordcloud library expects a string not a list, 
# so join the list back together with spaces
text = " ".join(all_lemmas)

# Generate a word cloud image
wordcloud = WordCloud().generate(text)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

# lower max_font_size
wordcloud = WordCloud(max_font_size=40).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
